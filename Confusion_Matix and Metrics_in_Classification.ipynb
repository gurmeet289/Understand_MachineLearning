## Confusion Matrix is a performance measurement for machine learning classification.

Okay!! now we can answer the small question...

What is Confusion Matrix and why you need it?

It is a performance measurement for machine learning classification problem where output can be two or more classes. It is a table with 4 different combinations of predicted and actual values.



It is extremely useful for measuring Recall, Precision, Specificity, Accuracy, and most importantly AUC-ROC curves.

Now, Lets understand TP, FP, FN, TN


As the easiest example to relate and understand is the pregnancy analogy-





### True Positive:

Interpretation: You predicted positive and its true.

You predicted that a woman is pregnant and she actually is.

### True Negative:

Interpretation: You predicted negative and its true.

You predicted that a man is not pregnant and he actually is not.

### False Positive: (Type 1 Error)

Interpretation: You predicted positive and its false.

You predicted that a man is pregnant but he actually is not.

### False Negative: (Type 2 Error)

Interpretation: You predicted negative and its false.

You predicted that a woman is not pregnant but she actually is.

## 7) F BETA OR F1 SCORE (F1 SCORE IS DERIVED FROM THE BETA VALUE)

### <font color='red'><b> F-betascore  = ((1+beta)*Precision*Recall)/beta*Precision+Recall <b></font>

### The F-beta score is the weighted harmonic mean of precision and recall, reaching its optimal value at 1 and its worst value at 0.

If beta= 1 it is F1-score ,

than, F-betascore  = ((2)*Precision*Recall)/Precision+Recall)-----> This is called Harmonic mean and if we replace,
Precision with (x) and Recall with (y)
than, the equation of F-betascore  = 2xy/x+y

When to use beta =1?
wherever, FP and FN both are equally important and both are having a greater impact at that time we can use beta = 1.

When to use beta =0.5?
When FP(Type 1 Error) having more impact than FN(Type 2 Error) than we reduce the beta value to 0.5 or beta =0.5

When to use beta =2?
When FN(Type 2 Error) having more impact than FP(Type 1 Error) than we increase the beta value to 2 or beta = 2

if beta=0.5 it is F0.5 score, if beta=2 it is F2 score
